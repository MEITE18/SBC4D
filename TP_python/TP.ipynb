```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_excel("D:\TP DEH ABOUBACAR\STAGE_SEJEN\MACHINE LEARNING\TP_regressionlineaire.xlsx")
print(dataset)
X = dataset.iloc[:, 1:2].values # Affecte X à toutes les colonnes sauf la dernière colonne
print(X)
Y = dataset.iloc[:, 4].values # Affecte Y à la dernière colonne.
print(Y)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
# Ajustement linéaire sur la donnée d'apprentissage et donnée test.
from sklearn.linear_model import LinearRegression
lrObj = LinearRegression()
lrObj.fit(X_train,Y_train)

# Prédiction sur la donnée test
Y_pred = lrObj.predict(X_test)
# Nous pouvons comparer les valeurs prédites avec les valeurs actuelles
print(Y_test)
print(Y_pred) # Valeurs prédites
```

```python
#Exploration visuelle de la donnée d'apprentissage
plt.scatter(X_train,Y_train,color='red')  # scatter pour indiquer des points et une couleur rouge.
plt.plot(X_train, lrObj.predict(X_train),color='blue') # plot pour indiquer la droite et une couleur bleue.
plt.title('Quantité de Co2 émise en fonction de la quantité de goudron contenue dans la cigarette on Training Set') # Donner un titre au graphique
plt.xlabel('TAR') # Titre de l'axe des abscisses
plt.ylabel('Quantité de CO2 émise')  # Titre de l'axe des ordonnées
plt.show() # Afficher le graphique
```

```python
#Exploration visuelle de la donnée test
plt.scatter(X_test,Y_test,color='yellow')  # scatter pour indiquer des points et une couleur jaune.
plt.plot(X_test, lrObj.predict(X_test),color='green') # plot pour indiquer la droite et une couleur verte.
plt.title('Quantité de Co2 émise en fonction de la quantité de goudron contenue dans la cigarette on Testing Set') # Donner un titre au graphique
plt.xlabel('TAR') # Titre de l'axe des abscisses
plt.ylabel('Quantité de CO2 émise')  # Titre de l'axe des ordonnées
plt.show() # Afficher le graphique
```

```python
# Evaluation de la performance du modèle
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error # On importe, le carré de RMSE, le R2 et le MAE.
# Si on veut calculer le R2 ajusté, on doit le faire nous même.
mean_squared_error(Y_test,Y_pred) # L'erreur moyenne par rapport à Y_test et Y_pred et qui est le carré de RMSE et s'interprète de la même manière.
```

```python
r2_score(Y_test, Y_pred)
# le TAR explique à 93% la quantité de CO2 émise. 
```

                *** Backward automatique ***

```python
# Nous avons besoin d'ajouter une colonne avec des 1 à X (correspondant au terme constant du modèle de regression)
X = np.append(arr=np.ones((24,1)).astype(int), values=X, axis=1) # ajouter 30 lignes de 1.
# Automatic backward selection
import statsmodels.formula.api as sm  
def backwardElimination(X,SL): 
    numVars =len(X[0]) 
    for i in range(0, numVars): 
        obj_OLS = sm.OLS(Y,X).fit() 
        maxVar = max(obj_OLS.pvalues).astype(float) 
        if maxVar > SL: 
            for j in range(0, numVars -i): 
                if (obj_OLS.pvalues[j].astype(float) == maxVar): 
                    X = np.delete(X, j,1)  
    obj_OLS.summary() 
    return X  
SL =0.05 
X_sig = X[:, [1, 2, 3]] 
X_Modeled = backwardElimination(X_sig, SL)

# type (X_modeled)
# print (X_modeled)
obj_OLS_final = sm.OLS(endog = Y, exog = X_Modeled).fit() 
obj_OLS_final.summary()
```
