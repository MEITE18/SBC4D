```python
#Elimination Backwardautomatique 
import statsmodels.formula.api as sm  
def backwardElimination(x,sl): 
    numVars =len(x[0]) 
    for i in range(0, numVars): 
        obj_OLS = sm.OLS(y,x).fit() 
        maxVar = max(obj_OLS.pvalues).astype(float) 
        if maxVar > sl: 
            for j in range(0, numVars -i): 
                if (obj_OLS.pvalues[j].astype(float) == maxVar): 
                    x = np.delete(x, j,1)  
    obj_OLS.summary() 
    return x  
SL =0.05 
x_sig = x[:, [0, 1, 2, 3, 4, 5]] 
x_Modeled = backwardElimination(x_sig,SL)

#tyoe(x_Modeled)
#print(x_Modeled)
obj_OLS.final=sm.OLS(endog=y, exog=x_Modeled).fit()
obj_OLS.final.summary()
```
