```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset= pd.read_csv("D:\TP DEH ABOUBACAR\STAGE_SEJEN\MACHINE LEARNING\salary.csv")
print(dataset)

x=dataset.iloc[:, 1:2].values #Prends toutes les colonnes sauf la derniere et affecte x toutes les colonnes sauf la derniere
y=dataset.iloc[:,-1].values # Affiche la derniere colonne et affecte Ã  y la derniere colone
print(x)
print(y)
```

```python
#Exploration visuelle de ladonnÃ©e  #dâapprentissage

#Ajustement lineaire regression sur la donnÃ©e d'apprentissage
from sklearn.linear_model import LinearRegression  
lrObj =LinearRegression() 
lrObj.fit(x,y)
plt.scatter(x,y,color='red')  # Affiche les points en rouge
plt.plot(x, lrObj.predict(x),color='blue')  # Affiche la droite en bleue
plt.title('Level vs Salary on Set')  
plt.xlabel('Level') 
plt.ylabel('Salary')  
plt.show()
```

```python
#PolynomialRegression 
from sklearn.preprocessing import PolynomialFeatures 
polyFeatureObj =PolynomialFeatures(degree=2) 
x_poly = polyFeatureObj.fit_transform(x)  
prObj = LinearRegression()  
prObj.fit(x_poly,y)

#Visualisation du resultat de la regression polynomiale
plt.scatter(x,y,color='red')  # Affiche les points en rouge
plt.plot(x, prObj.predict(x_poly),color='blue')  # Affiche la droite en bleue
plt.title('Resultats of Polynomial Regression Model')  
plt.xlabel('Level') 
plt.ylabel('Salary')  
plt.show()
```

```python
#PolynomialRegression 
from sklearn.preprocessing import PolynomialFeatures 
polyFeatureObj =PolynomialFeatures(degree=4) 
x_poly = polyFeatureObj.fit_transform(x)  
prObj = LinearRegression()  
prObj.fit(x_poly,y)

#Visualisation du resultat de la regression polynomiale
plt.scatter(x,y,color='red')  # Affiche les points en rouge
plt.plot(x, prObj.predict(x_poly),color='blue')  # Affiche la droite en bleue
plt.title('Resultats of Polynomial Regression Model')  
plt.xlabel('Level') 
plt.ylabel('Salary')  
plt.show()
```
